```python
df.rename(columns={"chromosome_name": "chromosome", "gene_lengh": "gene_length"}, inplace=True) #inplace =True modifies the original DataFrame
sarah = df.groupby( "chromosome")["gene_length"].sum()
sarah
```

```python

# Aggregate gene_length by chromosome and plot as bar
plt.bar(sarah.index, sarah.values, color="m") 
plt.title("Total Gene Length by Chromosome") 
plt.xlabel("Chromosome")
plt.ylabel("Total Gene Length")
plt.xticks(rotation= 90) 
for i, v in enumerate(sarah.values): 
    plt.text(i,v,v, ha='center', va='bottom') # Add value labels on top of bars
plt.show() #add this line to display the plot
```

```python
colors = [
    'tomato', 'cornflowerblue', 'gold', 'orchid', 'green',
    "#77BFE2"
]
plt.pie(sarah, startangle=90, labels=sarah.index, colors=colors*4)  # repeat colors if needed
plt.show()
```
